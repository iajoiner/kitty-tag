```python
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import requests
import pickle
import urllib.request
```

```python
#Global constant
URL = 'http://www.picato.net'
```

```python
# Process an individual cat pic page
# Input: the soup element produced by page_process
# Output: The dic containing info about the cat pic page
def ind_process(a_soup):
    ind_dict = {}
    zoom_href = a_soup['href']#The zoom page
    #print(zoom_href)
    req = requests.get(URL + zoom_href)
    req.raise_for_status()
    if req.status_code == requests.codes.ok:
        zoom_soup = BeautifulSoup(req.text, 'html.parser')
        #print(zoom_soup)
        ind_dict['url'] = zoom_soup.findAll('meta', {'property':'og:image'})[0]['content']
        tag_collection = zoom_soup.find('span',{'id':'PictureTags'})
        tags = tag_collection.findAll('a', {'class': 'hl BoxLH s nw'})
        #print(tag_collection)
        tag_clean = [tag.contents[0] for tag in tags]
        ind_dict['tags'] = tag_clean
        #print(ind_dict)
        return ind_dict
# Process a browse page
# Input: reduced url of the page
# Output: The list of page dicts and reduced URL of the earlier page
def page_process(page_href = '/browse/'):
    #print(URL + page_href)
    req = requests.get(URL + page_href)
    req.raise_for_status()
    if req.status_code == requests.codes.ok:
        soup = BeautifulSoup(req.text, 'html.parser')
        head = soup.findAll('a',{'class':'t ct'})
        tail = soup.findAll('a',{'class':'t '})
        head.extend(tail)
        page_dict_list = [ind_process(i) for i in head]
        return page_dict_list, next_page(soup)
# Scrape the entire site
# Output: The full dict list
def full_process(init_page_href = '/browse', output_file = '/Users/CatLover/Documents/DataScience/KittyTag/data/fdl.p'):
    page_href = init_page_href
    full_dict_list = []
    while page_href is not None:
        page_dict_list, page_href = page_process(page_href)
        full_dict_list.extend(page_dict_list)
        print(page_href)
        pickle.dump(full_dict_list, open(output_file,'wb'))
        copyfile(output_file, output_file + '.bac')
    return full_dict_list
# Produce the reduced URL of the earlier page
# Input: Soup of a browse page
# Output: Reduced URL of the earlier page if exists, otherwise produce None
def next_page(page_soup):
    eles = set()
    ele_list = page_soup(text='earlier')
    if len(ele_list) == 0:
        return None
    eles = {ele.parent for ele in ele_list}
    (ele,) = eles#Unpack, raise error if len(eles) != 1
    return ele['href']
def tag_extract(ind_page_href):
    req = requests.get(URL + ind_page_href)
    req.raise_for_status()
    if req.status_code == requests.codes.ok:
        soup = BeautifulSoup(req.text, 'html.parser')
        tag_collection = soup.find('span',{'id':'PictureTags'})
        tags = tag_collection.findAll('a', {'class': 'hl BoxLH s nw'})
        tag_clean = [tag.contents[0] for tag in tags]
        return tag_clean
```
